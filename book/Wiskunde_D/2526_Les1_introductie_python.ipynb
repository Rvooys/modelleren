# Structuur van code

Hieronder zie je een typisch voorbeeld van een python programma. Je hoeft de in en outs nog niet te kijken, we kijken eerst naar de zogeheten structuur van de code.

Een programma leest je code 'van boven naar beneden'. Je wilt dus vaak eerst beginnen met je 'beginwaarden', in programma hieronder zijn dat drie variabelen a, b en c, die de waarden 3, 5 en 7 willekeurig hebben gekregen.

Daarna wil je wat uitvoeren. Dat kan van alles zijn: 
- doe een controle, ALS een VOORWAARDE is voldaan, doe dan DIT:   (in het voorbeeld hieronder met if(a+b+c<20>): aangegeven)
- verander de waarden totdat aan een VOORWAARDE is voldaan (voorbeeld later: while)
- verander een aantal keer de waarden (voorbeeld later: for)

Met # kan je opmerkingen schrijven voor degene die je code gaat lezen, dan maak je het begrijpelijk wat je wilt doen of wat er gebeurd in die regel. In deze 'cursus' wordt dat vaak gedaan om nieuwe concepten toe te lichten. 

Lees de code hieronder en druk op run.


```python

a = 3   # met een hash-tag maak je opmerkingen
b = 5   #met 'a = 3' etc sla je de waarde 3 op in variabele a
c = 7
if (a+b+c<20):  #in het nederlands staat hier "als de som van a+b+c kleiner is dan 20, dan", je eindigt altijd met een :
    print("som is lager dan 20") #LET OP: hier komt een tabje, alles wat in dit blok wordt uitgevoerd tot het tabje verdwijnt
elif(a+b+c>=20): #als de som niet lager is, wat moet er dan gebeuren
    print("som is 20 of hoger") #let op het tabje

```

Van boven naar beneden:
- in drie regels worden 3 variabelen gedefinieerd. Bijvoorbeeld `a=3` betekend "de variabele a wordt gedefinieerd als getal 3". Je moet de =-teken dus NIET als vergelijking zien (zoals je wellicht gewend bent bij wiskunde), maar als DEFINITIE. 
- in regel 4 wordt een controle gedaan met een IF statement. 
- in regel 5 komt een TAB, dat een blok voorstelt, waarin de uitvoer wordt beschreven als aan de voorwaarde in regel 4 wordt voldaan. Omdat 3+5+7 inderdaad lager is dan 20, wordt met een commando print("...") een tekst geprint. (de output van de code). Op het print-commando komen we nog terug, voor nu is het genoeg om te weten dat het de output weergeeft.
- in regel 6 is de TAB weer terug, want je wilt niet dat dit uitgevoerd wordt als aan de vorige voorwaarde is voldaan. elif kan je in nederlands lezen als 'ZO NIET DAN ALS DIT', daarover later meer
- in regel 7 dus weer een tab vooruit, want er moet iets anders gebeuren.

*Getallen als variabelen*

Zoals je weet zijn er verschillende typen getallen: gehele getallen, negatieve getallen, breuken en irrationele getallen (zoals wortel(2)). Ook python maakt daar onderscheid in:
- gehele getallen heten integers, en worden afgekort als int
- decimale getallen heten floats
Python doet het best netjes, maar er zijn computertalen waarbij je fouten kan krijgen als je bijvoorbeeld een integer en een float wilt vermenigvuldigen. Krijg je een 'type error' dan is dat waarschijnlijk de fout. 
`a = 3 ` zorgt ervoor dat variabele a geforceerd een geheel getal blijft
`b = 3.0` zorgt ervoor dat variabele b geforceerd een decimaal is. 
Zie hieronder als voorbeeld, kijk naar de output. In de print() wordt drie keer een som berekend:
- a+2 = 5 want a en 2 zijn allebei integer
- a+2.0 = 5.0 want een int en een float geeft een float
- b+2 = 5.0 want een float en een int geeft een float
- a/2 en b/2 leveren allebei een float op, delen is een breuk dus een decimaal getal

```python
a = 4
b = 4.0

print("a+2=",a+2)
print("a+2.0=",a+2.0)
print("b+2=",b+2)
print(a/2)
print(b/2)
```

*Booleaanse variabelen* 

Booleaanse variabelen heb je vast nooit van gehoord, maar wordt veel gebruikt in programmeren. Er zijn twee waarden: `True` of `False`, ofwel waar of niet waar.  Let op de hoofdletters, die zijn verplicht in python.
Stiekem heb je al er eentje gezien in het eerste programma: `if(a+b+c>20)`. Het deel `a+b+c>20` neemt een van de twee waarden aan: als het klopt dat a+b+c groter dan 20 is, dan is het WAAR, en anders ONWAAR. De bewerkingen `>`, `<` of `==` of vormen daarvan initieren zo'n booleanse variabele.

# Enkele basisbewerkingen

Hier volgt een aantal zogeheten syntax regels, dwz regels hoe je bepaalde bewerkingen moet invoeren.

- Optellen en aftrekken is gewoon een + en een -
- Vermenigvuldigen gaat met een 'asterisk': *. Dus 5 keer 3 noteer je als `5*3`
- Delen gaat met een forward slash: / dus 15 gedeeld door 3 noteer je als `15/3`
- Machtsverheffen doe je met twee sterretjes: 5 tot de macht 3: `5**3`. De logica is dat machtsverheffen hetzelfde is als herhaaldelijk vermenigvuldigen: 5^3 = 5 keer 5 keer 5 dus driemaal met 5 vermenigvuldigen. 
- Eén enkele =-teken is in python "definitie", in andere programmeertalen kan dat ook `:=` zijn. Wil je juist testen of een variabele gelijk is aan een andere variabele, dan moet je twee =-tekens gebruiken. Bijvoorbeeld `if(D==0):` betekend 'Als D gelijk is aan 0, dan:'
- Groter gelijk of kleiner gelijk testen doe je met `>=` of `<=` .

Standaard drukt print() opdracht aan het einde op 'enter', het geeft in de output een nieuwe regel. 
Hieronder zie je een stukje code, waarin de eerste vijf kwadraten geprint worden (de for-loop wordt later uitgelegd!).

```python
for i in range(1,6):   # voor elke waarde i van 1 tot 6
    print(i**2)       # print i^2
```

Echter wil je soms de kwadraten op één regel hebben. Dat kan met de `end=""` optie, tussen de aanhalingstekens kan je ook andere symbolen gebruiken, bijvoorbeeld als output "1 en 4 en 9 en ..." wilt hebben:

```python
for i in range(1,6):
    print(i**2, end=" en ")
```

Let op dat er een extra "en" kom, omdat je tegen python zegt dat je een "en" moet printen aan het einde van je regel. 

Dit geeft mooi het verschil aan tussen de `sep` en de `end` optie:
- bij `sep` wordt TUSSEN elke output een object geplaatst (zoals spatie of en)
- bij `end` wordt NA elk output een object geplaatst

Hoe je dit oplost komt in een ander hoofdstuk aan bod

Een heel andere manier van opmaak is gebruik maken van een f-string. De letter f wordt herkend en je gebruikt accolades {} om variabelen in de tekst weer te geven. Zeker bij wiskunde is dat een fijne methode van werken voor de output, als je bijvoorbeeld niet alle 16 decimalen wilt hebben maar slechts 2. 

Voorbeelden zie je hieronder, met opmerkingen # is aangegeven wat er anders is en wat er gebeurd. In de basis geef je met `{variabele : opmaak}` aan welke variabele welke opmaak moet krijgen. 
Enkele opties:
.2f geeft getal in twee decimalen weer, de f staat voor 'fixed point'
.2e geeft het getal in twee decimalen in wetenschappelijke notatie (2.23e4 = 2.23x10^4 = 2230 )
een tabel maken, zie hieronder, met bijvoorbeeld .4d ga je 4 plekken opzij

```python
import math    # het begrip bibliotheek komt later
getal = math.pi   #sla de waarde pi op in getal
print(getal)   
print(f"Standaard: {getal}")  #zelfde output, je krijgt de eerste zoveel decimalen van pi
print(f"2 decimalen: {getal:.2f}")  #met getal:.2f geef je aan dat je een opmaak van 2 decimalen wilt
print(f"4 decimalen: {getal:.4f}")  #of 4 decimalen

grootgetal = 2342876341237
print(f"in wetenschappelijke notatie: {grootgetal:.5e}")  #het grote getal is in wetenschappelijke notatie afgerond op 5 decimalen


# een tabel 'tekenen' in een blok van 15 breed
#hieronder wordt een tabel getekend van x = 1 tot en met 3, en y = 10.5123*x
#de x-waarde wordt vanaf de 4e plek (.4d) gegeven (dimensie 4) en y vanaf positie 8, en worden met .2f twee decimalen afgerond
print("   x      y")
print("-" * 15) # Trek een lijntje van 15 strepen

for x in range(1, 4): # doe het volgende voor de waarden x=1, 2 en 3
    y = x * 10.5123   #bereken de waarde van y voor x=1, 2 en 3
    print(f"{x:4d} | {y:8.2f}")   #zie boven, tussen x waarden en y waarde komt een verticaal streepje
```

# De IF statement

Bij programmeren wil je wel eens tussendoor bepalen of aan een bepaalde conditie voldaan wordt, wat bepaald wat de volgende stap in de code gaat zijn.
Een bekend voorbeeld in de onderbouw is het algoritme om a x^2 + b x + c = 0 op te lossen: de ABC formule.

Daartoe moest de discriminant bepaald worden, en de waarde daarvan bepaalt het aantal oplossingen: Als D=0 is er één oplossing, als D>0 zijn er twee oplossingen en als D<0 zijn er geen oplossingen. 

Een van de opdrachten aan einde van de les wordt om de volledige abc formule te maken, voor nu bepalen we alleen het aantal oplossingen van de vergelijking 2x^2+3x-5=0



```python
a=2
b=3
c=-5
D=b**2 -4*a*c   #komt later, maar b**2 geeft het kwadraat van b
if(D>0):        #Als D groter is dan 0...
    print("er zijn twee oplossingen")
elif(D==0):     #Als D gelijk is aan nul, let op de twee =-tekens!
    print("er is één oplossing")
elif(D<0):      # Als D kleiner is dan nul...
    print("er zijn geen oplossingen")

print(D)
```

Opgevallen dat je twee =-tekens moet gebruiken? DAt komt omdat één =-teken betekend "definieer", `a=3` betekend definieer a als waarde 3.
Terwijl a==3 een gelijkheidscontrole is. Is de linkerkant gelijk gedefinieerd aan de rechterkant?

elif staat voor 'else if': "als de vorige voorwaarde niet voldaan is, controleer dan nu deze voorwaarde". Je kan willekeurig hoeveelheid elif blokken hebben.
Er bestaat ook iets als `else`, die gebruik je als aan geen van de voorwaarden voldaan is. Dit staat altijd aan het einde. Je komt in andere opdrachten wel een goed voorbeeld tegen.


We raken nu ook gelijk een andere type variabele aan die we nu kort bespreken: de booleanse variabele. Deze heeft twee waarden `True` of `False`.

`if (D>0)` wordt door python dus gelezen als `if (True)` dan gaat ie het blok na de dubbele punt volledig uitvoeren. Is het niet waar, dan gaat het naar de volgende if-statement.

# De For en While lus

Je bent hierboven al een paar keer het commando `for` tegen gekomen. Soms wil je een vast aantal berekeningen doen. In een van de voorbeelden wilden we de eerste vijf kwadraten weten, dat zijn 1^2 = 1, 2^2 = 4, 2^3 = 9 etc. 
In programma taal wil je dus voor elke waarde i, van i=1 tot en met i=4, het kwadraat uitrekenen en printen. 
In engels is dat "for i equal 1 to 5" en in programmeertaal is dat ineens `for i in range(1,6)`. 
Huh, waar komt die 6 vandaan? Het commando `range(a,b)` geeft een LIJST van de waarde a tot b (en niet tot en met)`. Het start met i=1, doet dan een kwadraat printen, en als het klaar is vervangt het de waarde i van 1 naar 2, etc, tot 6 bereikt is, dan gaan we 'uit de for-loop' springen en doen we wat anders.
Hieronder een voorbeeld voor de eerste 10 even getallen die we onder elkaar willen printen.


```python
for i in range(1,11):
    print(2*i)
```

Een for loop 'loopt' dus door een vaste set getallen heen, het heeft een duidelijk begin en een duidelijk einde. Je kan in range ook nog een stapgrootte aangeven met een 3e argument: `range(a,b,2)` slaat er steeds eentje over en zou je dus iets makkelijker de even getallen weer kunnen genereren. Bedenk zelf waarom voor 2 en 21 is gekozen

```python
for i in range(2,21,2):
    print(i)
```

Je kan ook 'tussentijds' uit de for loop springen met de break optie. Hier komt ook gelijk een optie van de if-statement naar voren.

```python
for i in range(1,100): #ga de getallen van 1 tot 100 af
    print(2*i) #bereken 2*i, zodat we een even getal hebben
    if(2*i>20):  #als de waarde van 2*i groter is dan 20, dan stappen we uit de for loop
        break
```

Een andere type lus is de while lus. Dat gebruik je als je berekeningen wil blijven uitvoeren tot een bepaalde voorwaarde is voldaan. Als je bijvoorbeeld niet weet hoeveel oneven getallen er onder de 20 zijn, kan je ook de while lus gebruiken:

```python
getal = 2     #definieer een start, 2 is het eerste even getal dus begin daar maar.
while getal < 21:  #we gaan door tot en met 20, dus tot 21
    print(getal)   #print het getal
    getal = getal + 2 #verhoog de waarde met 2
```

Een groot gevaar met de while lus is de zogeheten infinite loop: je blijft maar oneindig doorrekenen, omdat de eindconditie niet wordt bereikt.  Dat kan inherent aan de context zijn, bijvoorbeeld exponentieel verval dat nooit de waarde 0 bereikt. Zie hieronder. Vaak zal python dit herkennen en een foutmelding geven, maar het kan zijn dat je een paar uur moet wachten voordat het klaar is.
